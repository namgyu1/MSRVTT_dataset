```python
audio_file_path = sorted(glob.glob('./videos/audios_all/'+'*'))
```

```python
len(audio_file_path)
```

```python
import csv
test_vid = []

with open('./MSRVTT_JSFUSION_test.csv', newline='') as csvfile:
    data = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for i in data:
        test_vid.append(i[0].split(',')[2])
```

```python
test_audio_id = []
for file in audio_file_path:
    f_name = file.split('/')[-1][:-4]
    if str(f_name) in test_vid:
        test_audio_id.append(f_name)
```

```python
test_no_audio_id = []
for file in test_vid:
    if file not in test_audio_id:
        test_no_audio_id.append(file)
```

```python
print("There are {} videos which have audio information and {} videos which do not have within total 1000 videos".format(len(test_audio_id), len(test_no_audio_id)))
```
